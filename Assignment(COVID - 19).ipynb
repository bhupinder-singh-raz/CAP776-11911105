# 1. World wide Confirmed Cases

```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.offsetbox import OffsetImage,AnnotationBbox
df=pd.read_excel('covid_world.xlsx')
df1=df
```

```python
c=df1.head(12)
```

# 2. India Covid-19 Cases Statewise

```python
df=pd.read_excel('CovidStateWise.xlsx')
df2=df
```

```python
#df2
```

```python
confirmed=df2["Confirmed"].tolist()

plt.figure(figsize=(25,25))

def value(coord, name, ax):
    xaxis=confirmed[i]
    #print(xaxis)
    plt.text(xaxis + 10 , i - .2 , xaxis , fontsize=20)

plt.barh(df2['State'],df2['Confirmed'],color='#F4D03F',label='Confirmed',edgecolor='blue')
plt.barh(df2['State'],df2['Recovered'],color='#52BE80',label='Recoverd')
plt.barh(df2['State'],df2['Deaths'],color='#E74C3C',label='Deaths')

plt.title('India Covid-19 StateWise Cases till 12 April',fontsize=30,pad=45 )
plt.ylabel('States',fontsize=40,color='#3498DB')
plt.xlabel('Total No. of Cases',fontsize=40,color='#3498DB')
plt.yticks(df2['State'],fontsize=20)
plt.xticks([0,200,400,600,800,1200,1400],fontsize=20)
plt.legend(fontsize=20)


for i, c in enumerate(df2['State']):
    value(i,c,ax)

plt.show()
```

It is noticed from the bar graph that 1364 Confirmed cases are found in the Maharashtra 

And States such as Sikkim , Nagaland , Meghalaya , Daman and Diu still have no cases yet . 

# 3. India Covid-19 Cases AgeWise

```python
import pandas as pd
from matplotlib import pyplot as plt
df=pd.read_csv('AgeGroupDetails.csv')
df3=df.head(9)
```

```python
#df3
```

```python
plt.figure(figsize=(9,8))

colors = ['#1D8348','#2ECC63','#4CFD8F','#F7DC6F','#F4D03F','#F1C40F','#FD8A4C','#E74C3C','#A93226']
labels = ['over 80s','Age (0 - 9)','Age (10 - 19)' , ' Age (70 - 79)' ,'Age (50 - 59)' ,'Age (60 - 69)' ,'Age (40 - 49)','Age (30 - 39)','Age (20 - 29)']
explode = (0.2,0, 0, 0 , 0 , 0 , 0 , 0 , 0.1)

Tc = df3['TotalCases'].sort_values()
plt.pie(Tc,explode=explode,labels=labels,colors=colors, autopct='%1.1f%%', shadow=True)

plt.title('India Covid-19 Cases AgeWise',fontsize=28,pad=42 )
plt.show()
```

# Obserations

It is observed from the pie chart that Adults and Teenagers have more cases. Alomst one quartle of chart

On the other hand children and old people have least no of cases.
